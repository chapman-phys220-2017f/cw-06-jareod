```python
import numpy as np
import array_calculus as ac
```

# Implementation of Code

### Plotting $x^2$

First, we examine our code acting on $f(x) = x^2$.

```python
a = -5
b = 5
n = 1000
ac.plot_function(a,b,n,ac.f(a,b,n),"x**2")
```

We see what we would expect.  The blue plot represents our $f(x) = x^2$, and is quadratic in shape.  The red plot is the first derivative and is linear, which would be expected because we know $f'(x)=2x$.  The green plot is the second derivative and is constant minus a glitch at both ends of the domain.  We would expect the second derivative to be constant such that $f''(x)=2$, which it is.

### Plotting $\sin(x)$

Next, we examine our code on the function $s(x) = \sin(x)$

```python
a = -2*np.pi
b = 2*np.pi
n = 1000
ac.plot_function(a,b,n,ac.s(a,b,n),"sin(x)")
```

We see what is expected.  The blue plot is the function, $s(x) = \sin(x)$.  We know the red plot is the first derivative, $s'(x) = \cos(x)$, so $s'(0) = 1$, in theory.  This lines up with our plot.  The second derivative is represented by the green plot.  We know $s''(x) = -\sin(x)$, so $s''(0) = 0$.  Once again, this is consistent with our plot.

### Plotting the Gaussian

Finally, we implement our code on the Gaussian function.

```python
a = -4
b = 4
n = 100
ac.plot_function(a,b,n,ac.g(a,b,n),"Gaussian")
```

The plot that we see makes sense.  The blue plot is, obviously, the Gaussian.  The red plot is the first derivative.  We see that the zero of the first derivative lines up with the maximum of the Gaussian, as we would expect.  The green plot is the second derivative of the Gaussian.  This plot seems to fit the mold as well.

Our code for a 2-dimensional array to model an approximate derivative funciton seems to have been successful for 3 seperate functions, the square, sine, and Gaussian.
